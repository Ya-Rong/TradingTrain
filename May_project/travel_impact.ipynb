```python
import requests

API_KEY = 'AIzaSyCVU2e05B_wQNADFRDv06gH7wLP9HhGjuU'
url = 'https://travelimpactmodel.googleapis.com/v1/flights:computeFlightEmissions?key=' + API_KEY

headers = {'Content-Type': 'application/json'}
data = {
    "flights": [
        {
            "origin": "ZRH",
            "destination": "CDG",
            "operating_carrier_code": "AF",
            "flight_number": 1115,
            "departure_date": {"year": 2023, "month": 11, "day": 1}
        },
        {
            "origin": "CDG",
            "destination": "BOS",
            "operating_carrier_code": "AF",
            "flight_number": 334,
            "departure_date": {"year": 2023, "month": 11, "day": 1}
        },
        {
            "origin": "ZRH",
            "destination": "BOS",
            "operating_carrier_code": "LX",
            "flight_number": 54,
            "departure_date": {"year": 2023, "month": 7, "day": 1}
        }
    ]
}

response = requests.post(url, headers=headers, json=data)
# print(response.content.decode('utf-8'))

response_data = response.json()
print(response_data)

# 解析回應資料
for flight_result in response_data['flight_emissions']:
    origin = flight_result['origin']
    destination = flight_result['destination']
    operating_carrier_code = flight_result['operating_carrier_code']
    flight_number = flight_result['flight_number']
    emissions = flight_result['emissions']

    # 提取航班詳細資料
    flight_details = flight_result['flight_details']
    departure_time = flight_details['departure_time']
    arrival_time = flight_details['arrival_time']
    price = flight_details['price']

    print('Flight:', operating_carrier_code, flight_number)
    print('Origin:', origin)
    print('Destination:', destination)
    print('Departure Time:', departure_time)
    print('Arrival Time:', arrival_time)
    print('Price:', price)
    print('---')
```
