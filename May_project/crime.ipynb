```python
import requests
import json

url = "https://od.moi.gov.tw/api/v1/rest/datastore/A01010000C-000499-193"

# 下載檔案
response = requests.get(url)
data = response.json()

# 解析資料
result = data["result"]
records = result["records"]

# 輸出解開的資料
for record in records:
    print(record)
```

```python
# 建立用於分類的字典
classified_data = {}

# 將資料分類
for record in records:
    record_type = record["type"]
    if record_type in classified_data:
        classified_data[record_type].append(record)
    else:
        classified_data[record_type] = [record]

# 輸出分類結果
for record_type, records in classified_data.items():
    print(f"Type: {record_type}")
    for record in records:
        print(record)
    print()
```

```python
# 建立用於計算次數的字典
type_counts = {}

# 計算次數
target_type = '機車竊盜'

for record in records:
    record_type = record["type"]
    if record_type == target_type:
        if target_type in type_counts:
            type_counts[target_type] += 1
        else:
            type_counts[target_type] = 1

# 輸出計算結果
count = type_counts.get(target_type, 0)
print(f"Type: {target_type}, Count: {count}")
```
